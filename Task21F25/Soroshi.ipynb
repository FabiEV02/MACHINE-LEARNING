##**Problem 1: Number of grains of rice on day 100**
1.   Day1: 1
2.   Day2: 1 x 2 = 2
3.   Day3: 1x2 + 1x2 = 4
4.   Day4: 1x2 + 1x2 + 1x2 + 1x2 = 8

This sequence can be summarized as $2^n$ where n represents (day -1)

We are going to use the for loop that will iterate from i=0 (inclusive) to i=100 (exclusive), in each iteration the operation $2^i$ is performed and it is added to a list, in addition this value is accumulated in the variable total. As a result we will have a list with the number of rice per day and the total number of rices after 100 days. This data will be printed on the console and graphed.

```python
import matplotlib.pyplot as plt
%matplotlib inline

list = []
total = 0
for i in range(100):
  list.append(2**i)
  total = total + 2**i
print(f"Number of grains of rice on the 100th day:  {list[-1]}")
print(f"The total rice grain after 100 days: {total}")
days = range(1,101)
plt.title("Number of rice grains", fontsize=16, color='red')
plt.xlabel("number of days", fontsize=14, color='blue')
plt.ylabel("number of grains of rice", fontsize=14, color='blue')
plt.plot(days,list, color='hotpink', linewidth=3, linestyle='--')
plt.grid(True, linestyle=':', color='gray', linewidth=0.8)
plt.gca().set_facecolor('lightyellow')
plt.show()
```

##Problem 2: Number of grains of rice other than 100 days

In this problem the number n of days is entered by keyboard, n is the argument of a function that creates a list with the amount of rice per day from day 1 to day n and also accumulates the total of rice in the variable total, a list is also created that contains the values ​​from 1-n.
The function returns the list with the amount of rice per day, the list of days (1-n) and the variable total.
Finally, with all this, the graph is made and printed in the console.

```python
import matplotlib.pyplot as plt
%matplotlib inline
lista = []
total = 0

def compute_sorori_shinzaemon(n):
  total = 0
  days = range(1, n+1)
  for i in range(n):
    lista.append(2**i)
    total = total + 2**i
  return lista,days,total

n = int(input("Day: "))
lista,days,total = compute_sorori_shinzaemon(n)
print(f"Number of grains of rice on the day {n}:  {lista[-1]}")
print(f"The total rice grain after days {n}: {total}")
plt.title("Number of rice grains", fontsize=16, color='red')
plt.xlabel("number of days", fontsize=14, color='blue')
plt.ylabel("number of grains of rice", fontsize=14, color='blue')
plt.plot(days,lista, color='hotpink', linewidth=3, linestyle='--')
plt.grid(True, linestyle=':', color='gray', linewidth=0.8)
plt.gca().set_facecolor('lightyellow')
plt.show()
```

##Problem 3: How many people can live how many days with the rice they receive?

Taking into account that the total amount of rice is $2^{99}$ and that on average a person consumes 4000 grains of rice per day, we can say that:
- The total amount of rice is enough for: $ \frac{2^{99}}{4000} $ days
- Finally, I divide that number of days by the number of people in the class
In the program, the number of people is entered by keyboard and the calculator function, which has the number of people and the total amount of rice as arguments, calculates how many days each person will receive rice, returns this value and prints it on the console.

```python
def calculator(ngrains,npeople):
  #Each person consumes 4000 grains of rice per day
  ndays = ngrains/4000
  dayslife = ndays/npeople
  return dayslife

ngrains = 2**99
npeople = int(input("Number of people in the class: "))
dayslife = calculator(ngrains,npeople)
print(f"Every student can live {dayslife} days")
```
