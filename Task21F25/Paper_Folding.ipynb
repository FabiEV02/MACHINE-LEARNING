##**Problem 1: Implementation using arithmetic operators for powers.**
Create a program that uses exponentiation arithmetic operators to calculate the thickness of a sheet of paper when folded 43 times.

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 **43
print(f"Thickness: {folded_thickness} meters")
```

##**Problem 2: Unit conversion**
Convert meters to kilometers with two decimal places.

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 **43
print("Thickness: {:.2f} kilometers".format(folded_thickness / 1000))
```

##**Problem 3: Created using a for statement**
Create a program that uses the for statement to calculate the thickness of a sheet of paper when folded 43 times.
Do not use exponentiation arithmetic operators.

```python
# range(43) generates numbers from 0 to 42
THICKNESS = 0.00008
for i in range(43):
  THICKNESS *=2
print("Thickness: {} meters".format(THICKNESS))
print("Thickness: {:.2f} kilometers".format(THICKNESS / 1000))
```

##**Problem 4: Comparison of calculation times**

```python
import time
THICKNESS = 0.00008
#Programa 1
start1 = time.time()
folded_thickness = THICKNESS * 2 **43
elapsed_time1 = time.time() - start1
#Programa 2
start2 = time.time()
for i in range(43):
  THICKNESS *=2
elapsed_time2 = time.time() - start2

dif = elapsed_time1 - elapsed_time2
if dif > 0:
  print(f"The calculation performed with the for statement was faster and the difference between both methods was:{dif:.10f} [s]")
else:
  print(f"The calculation performed with the power operator was faster and the difference between both methods was:{-1*dif:.10f} [s]")
```

##**Problem 5: Save to list**
*   First, create an empty list.
*   The prefolded value is then added to the list.
*   The values ​​from the process of folding the paper up to n times in a for statement are added to the list.

Finally, it is important to check that there are 44 values ​​stored in the list using the function len.


```python
THICKNESS = 0.00008
list = []
list.append(THICKNESS)
for i in range(43):
  THICKNESS *=2
  list.append(THICKNESS)
print(f"List length: {len(list)}")
```

##**Problem 6: Visualizing a linear graph**
Draw a graph Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(list)
plt.show()
```
